```python
%config InlineBackend.figure_formats = ['pdf']
import numpy as np
import matplotlib.pyplot as plt
from ckine.figures.figureB1 import runIL2simple, IC50global, numpy_data, cell_names, rxntfR
```

```python
sampleRange = np.logspace(-1.0, 2.0, num=100)
EChelp = np.array([IC50global([1.0, x, 5.0], input_receptors=numpy_data[7, 0:3]) for x in sampleRange])
ECNK =   np.array([IC50global([1.0, x, 5.0], input_receptors=numpy_data[0, 0:3]) for x in sampleRange])
ECreg =  np.array([IC50global([1.0, x, 5.0], input_receptors=numpy_data[4, 0:3]) for x in sampleRange])
EChelpRed = np.array([IC50global([8.0, x, 5.0], input_receptors=numpy_data[7, 0:3]) for x in sampleRange])
ECNKRed =   np.array([IC50global([8.0, x, 5.0], input_receptors=numpy_data[0, 0:3]) for x in sampleRange])
ECregRed =  np.array([IC50global([8.0, x, 5.0], input_receptors=numpy_data[4, 0:3]) for x in sampleRange])

plt.semilogx(1.0 / sampleRange, EChelp / ECreg, label="Treg / Thelper", color='C0');
plt.semilogx(1.0 / sampleRange, ECNK / ECreg, label="Treg / NK", color='C1');
plt.semilogx(1.0 / sampleRange, EChelpRed / ECregRed, label="Treg / Thelper", color='C0', ls=':');
plt.semilogx(1.0 / sampleRange, ECNKRed / ECregRed, label="Treg / NK", color='C1', ls=':');
plt.xlabel("Relative CD122/CD132 affinity");
plt.ylabel("Specificity");
plt.legend();
plt.ylim(bottom=0.0);
```

```python
plt.loglog(1.0 / sampleRange, ECreg, color='C0');
plt.loglog(1.0 / sampleRange, ECregRed, color='C0', ls=':');
plt.xlabel("Relative CD122/CD132 affinity");
plt.ylabel('$T_{reg} \;\; EC_{50}$');
plt.legend();
plt.ylim(top=10, bottom=0.001);
```
