### Calculating error

```python
IL2_sum_sq_err = np.sum((mat[:,34:50])**2, axis=1)
IL15_sum_sq_err = np.sum((mat[:,50:66])**2, axis=1)
traf_sum_sq_err = np.sum((mat[:,66:94])**2, axis=1)

iteration = np.arange(1,501)
IL2_samples = 16
IL15_samples = 16
traf_samples = 28
```

#### Plot sum of squared error for all three measurements

```python
plt.scatter(iteration, IL2_sum_sq_err, s=2, color='r', label='IL2', alpha=0.5)
plt.scatter(iteration, IL15_sum_sq_err, s=2, color='c', label='IL15', alpha=0.5)
plt.scatter(iteration, traf_sum_sq_err, s=2, color='g', label='traf', alpha=0.5)
plt.legend()
plt.title('Sum of squared error')
plt.xlabel('iteration')
plt.ylabel('sum squared error')
plt.show()

print('average sum squared error for IL2 data is ' +str(np.mean(IL2_sum_sq_err)))
print('average sum squared error for IL15 data is ' +str(np.mean(IL15_sum_sq_err)))
print('average sum squared error for trafficking data is ' +str(np.mean(traf_sum_sq_err)))
```

#### Plot mean squared error for all three measurements

```python
plt.scatter(iteration, (IL2_sum_sq_err/IL2_samples), s=2, color='r', label='IL2', alpha=0.5)
plt.scatter(iteration, (IL15_sum_sq_err/IL15_samples), s=2, color='c', label='IL15', alpha=0.5)
plt.scatter(iteration, (traf_sum_sq_err/traf_samples), s=2, color='g', label='traf', alpha=0.5)
plt.legend()
plt.title('Mean squared error')
plt.xlabel('iteration')
plt.ylabel('mean squared error')
plt.show()

print('average mean squared error for IL2 data is ' +str(np.mean(IL2_sum_sq_err) / IL2_samples))
print('average mean squared error for IL15 data is ' +str(np.mean(IL15_sum_sq_err) / IL15_samples))
print('average mean squared error for trafficking data is ' +str(np.mean(traf_sum_sq_err) / traf_samples))
```
